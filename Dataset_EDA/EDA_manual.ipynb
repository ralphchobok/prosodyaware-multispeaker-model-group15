```python
# Reload CSV without headers
df = pd.read_csv('/home/ralph_c/ralph/lightspeech/modern_fastspeech_lightspeech/fast_speech_english_customized/LJSpeech-1.1/metadata.csv', 
                 sep='|', 
                 header=None, 
                 names=['id', 'text', 'normalized_text'])

print("Column names:", df.columns.tolist())
print(f"\nTotal rows: {len(df)}")
print(f"\nFirst few rows:")
print(df.head())
```

```python
# Calculate average number of words in the second column (text)
df['word_count'] = df['text'].str.split().str.len()

average_words = df['word_count'].mean()
print(f"Average number of words in the text column: {average_words:.2f}")
```

```python
# Find the smallest text (by word count)
min_idx = df['word_count'].idxmin()
smallest_text = df.loc[min_idx]

print(f"Smallest text (by word count):")
print(f"ID: {smallest_text['id']}")
print(f"Word count: {smallest_text['word_count']}")
print(f"Text: {smallest_text['text']}")
print(f"\n{'-'*80}\n")
```

```python
# Find the longest text (by word count)
max_idx = df['word_count'].idxmax()
longest_text = df.loc[max_idx]

print(f"Longest text (by word count):")
print(f"ID: {longest_text['id']}")
print(f"Word count: {longest_text['word_count']}")
print(f"Text: {longest_text['text']}")
```
